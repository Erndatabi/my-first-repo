```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# 1. 定義抓取清單：標普 500 (SPY), 黃金 (GLD), 十年期美債收益率 (^TNX)
symbols = ['SPY', 'GLD', '^TNX']
print("正在從 Yahoo Finance 抓取數據...")

# 2. 抓取過去一年的數據
data = yf.download(symbols, period='1y')['Close']

# 3. 數據標準化 (Normalization)
# 因為指數係 5000 幾點，金價係 2000 幾點，直接畫會睇唔到
# 我哋將起點全部設為 100，咁就睇到邊份資產跑贏
norm_data = (data / data.iloc[0]) * 100

# 4. 畫圖
norm_data.plot(figsize=(12, 6))
plt.title('Cross-Asset Performance (Last 1 Year)')
plt.ylabel('Normalized Price (Start = 100)')
plt.grid(True)
plt.show()

# 5. 睇埋相關性 (Correlation)
print("\n資產相關性矩陣：")
print(data.corr())
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

symbols = ['SPY', 'GLD', '^TNX']
# 確保數據係對齊嘅 (dropna 刪除唔完整嘅日期)
data = yf.download(symbols, period='1y')['Close'].dropna()

# 關鍵：我哋要確保 norm_data 入面真係有三個 Columns
norm_data = (data / data.iloc[0]) * 100

# 檢查一下數據前幾行，睇吓係咪三列都有數字
print(norm_data.head())

# 畫圖時強制畫出所有 Line
fig, ax = plt.subplots(figsize=(12, 6))
norm_data.plot(ax=ax)

plt.title('Cross-Asset Performance (Standardized to 100)')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```

```python
# 範例：左邊睇 SPY 價格，右邊睇 Yield 變動
fig, ax1 = plt.subplots(figsize=(12, 6))

ax1.set_xlabel('Date')
ax1.set_ylabel('SPY Price', color='tab:blue')
ax1.plot(data.index, data['SPY'], color='tab:blue')

ax2 = ax1.twinx() # 建立第二個 Y 軸
ax2.set_ylabel('10Y Yield (%)', color='tab:green')
ax2.plot(data.index, data['^TNX'], color='tab:green')

plt.show()
```

```python
# 畫完圖之後，將圖表儲存為 PNG 檔案
plt.savefig('cross_asset_chart.png', dpi=300, bbox_inches='tight')
print("圖表已成功儲存為 cross_asset_chart.png")
```
